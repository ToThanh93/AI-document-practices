```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, adjusted_rand_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.manifold import TSNE
import skfuzzy as fuzz
import pandas as pd
import matplotlib.pyplot as plt
import umap
import numpy as np

# Đọc dữ liệu
data = pd.read_csv('university.data', delimiter="\t")

# Tiền xử lý dữ liệu
for col in data.columns:
    if data[col].dtype == 'object':
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col])

# Chuẩn hóa dữ liệu
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Phân cụm bằng K-means
kmeans = KMeans(n_clusters=3, random_state=0).fit(data_scaled)

# Lấy nhãn cụm
kmeans_labels = kmeans.labels_

# Tính ARI
true_labels = data['(def-instance Adelphi']
ari_kmeans = adjusted_rand_score(true_labels, kmeans.labels_)

# Giảm chiều dữ liệu với t-SNE
tsne = TSNE(n_components=2, random_state=0)
data_tsne = tsne.fit_transform(data_scaled)

# Giảm chiều dữ liệu với UMAP
reducer = umap.UMAP()
data_umap = reducer.fit_transform(data_scaled)

# Trực quan hóa kết quả phân cụm
fig, ax = plt.subplots(1, 2, figsize=(15, 7))

for axi, data_reduced, title in zip(ax, [data_tsne, data_umap], ['t-SNE', 'UMAP']):
    # Trực quan hóa dữ liệu
    scatter = axi.scatter(data_reduced[:, 0], data_reduced[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.5)
    # Trực quan hóa centroid nếu có
    if hasattr(kmeans, 'cluster_centers_'):
        centers_reduced = tsne.fit_transform(kmeans.cluster_centers_)
        axi.scatter(centers_reduced[:, 0], centers_reduced[:, 1], c='red')
    axi.set_title(title)
    axi.legend(*scatter.legend_elements(), title="Clusters")

plt.show()

# Thêm nhãn cụm vào dữ liệu
data['cluster_label'] = kmeans.labels_

# Tạo hồ sơ cho mỗi cụm
cluster_profiles = data.groupby('cluster_label').mean()
print(cluster_profiles)
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
import skfuzzy as fuzz
import pandas as pd
from scipy.spatial.distance import cdist

# Đọc dữ liệu
data = pd.read_csv('crx.data')

# Chuẩn bị dữ liệu (chuyển đổi dữ liệu dạng chuỗi thành số, xử lý giá trị NaN, v.v.)

# Phân cụm bằng K-means
kmeans = KMeans(n_clusters=3, random_state=0).fit(data)
kmeans_labels = kmeans.labels_

# Phân cụm bằng Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=3).fit(data)
agg_labels = agg.labels_

# Phân cụm bằng DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5).fit(data)
dbscan_labels = dbscan.labels_

# Phân cụm bằng Fuzzy C-means
cntr, u, _, _, _, _, _ = fuzz.cluster.cmeans(data.T, 3, 2, error=0.005, maxiter=1000)
fcm_labels = np.argmax(u, axis=0)

# Tính chỉ số Silhouette
silhouette_avg_kmeans = silhouette_score(data, kmeans_labels)
silhouette_avg_agg = silhouette_score(data, agg_labels)
silhouette_avg_dbscan = silhouette_score(data, dbscan_labels)
silhouette_avg_fcm = silhouette_score(data, fcm_labels)

print(f"Silhouette Score is : {silhouette_avg_kmeans} (K-means), {silhouette_avg_agg} (Agglomerative), {silhouette_avg_dbscan} (DBSCAN), {silhouette_avg_fcm} (Fuzzy C-means)")

# Sử dụng phương pháp Elbow để xác định số lượng cụm tối ưu
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(data)
    distortions.append(sum(np.min(cdist(data, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / data.shape[0])

# Vẽ đồ thị Elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```
